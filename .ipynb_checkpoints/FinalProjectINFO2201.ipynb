```python
#import pandas
import pandas as pd
#create a data frame of our csv file
df = pd.read_csv('number-species-threatened/number-species-threatened.csv')
#show the data frame to see the data
df
```

```python
# to do a stacked bar chart make a copy of the data frame so that we can use said "new" data frame to make the 
#chart based off of that 
copy_main_df = df.copy()

# Drop the rows we don't really need- we aren't getting much value out of them
copy_main_df = copy_main_df.drop([0,1,2,3,24])

# Drop the year column

#instead of deleting the code column, change what they're classified as to be either invertebrates, fungi, vertebrates, or plants
copy_main_df['Code'] = ['vertebrates', 
                        'invertebrates', 
                        'vertebrates',
                        'invertebrates',
                        'invertebrates',
                        'vertebrates',
                        'plant',
                        'vertebrates',
                        'plant',
                        'plant',
                        'plant',
                        'invertebrates',
                        'invertebrates',
                        'vertebrates',
                        'invertebrates',
                        'plant',
                        'plant',
                        'invertebrates',
                        'plant',
                        'vertebrates',
                        'invertebrates'
                       ]


# make a pivot table, this will help for later
pivot_df = copy_main_df.pivot_table(index = "Code", columns="Entity", values="Number of threatened species", fill_value=0)
pivot_df
```

```python
#Cleaning the data by getting rid of columns that dont necessarily help our story telling. 
del df['Code']
```

```python
#show the new table
df
```

```python
#Created another dataframe using .drop to exclude the species we won't use. 
#The clean_df prints out the new chart 
clean_df = df.drop([0,1,2,3,5,7,13,18,19,16,20,21,22,24,25])
clean_df = clean_df.reset_index(drop = True)

clean_df
```

```python
#importing seaborn so we can make our data visualizations
import seaborn as sns
```

```python
#Created a plants_df only using plant species.
plants_df = clean_df.drop([0,1,2,3,5,8,9,10])
plants_df = plants_df.reset_index(drop = True)

plants_df
```

```python
#The terrestrial_df are the 
terrestrial_df = clean_df.drop([2,3,4,5,6,7,8])
terrestrial_df = terrestrial_df.reset_index(drop = True)

terrestrial_df
```

```python
#The aquatic_df will print the aquatic and semi-aquatic creatures 
aquatic_df = clean_df.drop([1,4,6,7,9])
aquatic_df = aquatic_df.reset_index(drop = True)

aquatic_df
```

```python
#DATA VISUALIZATIONS START HERE:
#making all of our data visualizations from this point on. Everything above was to clean the data to get us to
#this point.
```

```python
p = aquatic_df.plot(
    kind='bar',
    x="Entity",
    y="Number of threatened species",
    legend = False,
    title = 'Number Of Threatened Species',
    color = ['skyblue','aqua','skyblue','purple','lightgreen','lightgreen']
)
p
```

```python
p = plants_df.plot(
    kind='bar',
    x="Entity",
    y="Number of threatened species",
    legend = False,
    title = 'Number Of Threatened Species',
    color = ['lightgreen','darkgreen','limegreen','purple','lightgreen','lightgreen']
)
p
```

```python
p = terrestrial_df.plot(
    kind='bar',
    x="Entity",
    y="Number of threatened species",
    legend = False,
    title = 'Number Of Threatened Species',
    color = ['skyblue','aqua','skyblue','purple','lightgreen','lightgreen']
)
p
```

```python
#Created a combined dataset that totals the threatened plant species
plants_sum = sum(plants_df['Number of threatened species'])
plants_sum
```

```python
#Created a combined dataset that totals the threatened terrestrial species
terrestrial_sum = sum(terrestrial_df['Number of threatened species'])
terrestrial_sum
```

```python
#Created a combined dataset that totals the aquatic plant species
aquatic_sum = sum(aquatic_df['Number of threatened species'])
aquatic_sum
```

```python
#Made new column names 
column_names = ['Entity Type','Quantity']
#Created a dictionary with the totaled dataframes. 
data = {}
data['Entity Type'] = ['Aquatic', 'Terrestrial', 'Plants']  
data['Quantity'] = [aquatic_sum, terrestrial_sum, plants_sum] 
summary_df = pd.DataFrame(data, columns=column_names)
summary_df
```

```python
#Import matplotlib.pyplot to help with the pie chart
import matplotlib.pyplot as plt
```

```python
#Made a summary_df pie chart from matplot with the quantities converted to percentages. Also, customized the colors in the pie chart. 
label_counts = summary_df['Quantity']
plt.pie(label_counts,
        labels = summary_df['Entity Type'],
        autopct = '%1.1f%%',
        colors = ['aqua','brown','lightgreen'],
       )
plt.show()
```

```python
# Stacked bar chart using the copy_main_df we created earlier 
p = pivot_df.plot(kind='bar',
                  stacked=True,
                  legend=False)
plt.xlabel('Species Type')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
```
